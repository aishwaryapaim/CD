```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from io import StringIO
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
# create webdriver object
ser = Service("D:/Docs/chromedriver.exe")
op = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=ser, options=op)
#driver = webdriver.Chrome('D:/Docs/chromedriver.exe')vbm ,n.
driver.maximize_window()  
# get google.co.in
driver.get("https://www.pcs.agriculture.gov.ie/pest.asp?searchType=active")
time.sleep(1)
r = driver.find_elements(by=By.XPATH, value='//*[@id="activeSub"]/option')
print(len(r))
```

```python
Chemical_df = pd.DataFrame(columns = ['Product Name','PCS No.','Authorization Holder','Marketing Company','Product Type','Function','User Type'])
ActiveSubstance_df = pd.DataFrame(columns = ['PCS No.','Active Substance','Content'])


for i in range(10):
    driver.get("https://www.pcs.agriculture.gov.ie/pest.asp?searchType=active")
    driver.find_elements(by=By.XPATH, value='//*[@id="activeSub"]/option')[i].click()
    driver.find_element(By.ID, "getActive").click()
    driver.find_element(by=By.XPATH, value='//div[@id="content"]//table//div').click()
    strUrl = driver.current_url
    tables = pd.read_html(strUrl)
    df1 = tables[3]
    df2 = tables[4]
    df1 = df1.transpose()
    df1.reset_index(drop=True, inplace=True)
    df1.columns = df1.iloc[0]
    df1 = df1[1:]
    df2.columns = df2.iloc[0]
    df2 = df2[1:]
    df2.insert(0, 'PCS No.', df1.iloc[0][1])
    frames = [Chemical_df, df1]
    Chemical_df = pd.concat(frames)
    frames = [ActiveSubstance_df, df2]
    ActiveSubstance_df = pd.concat(frames)

    
```

```python
ActiveSubstance_df.drop_duplicates(keep='first',inplace=True)
Chemical_df.drop_duplicates(keep='first',inplace=True)
```
